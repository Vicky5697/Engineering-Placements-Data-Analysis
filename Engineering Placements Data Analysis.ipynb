### Performing EDA

#### Gender Count

```python
plt.figure(figsize=(10,5))
sns.countplot(x=data['Gender'])
plt.xlabel('Gender', fontsize=(10))
plt.ylabel('Count')
plt.title('Gender Count')
plt.show()
```

```python
plt.figure(figsize=(10,5))
labels = ['Males','Female']
count = [len(data[data['Gender']=='Male']), len(data[data['Gender']=='Female'])]
plt.pie(count, labels=labels, autopct='%1.1f%%')
plt.title('Gender Percentage')
plt.show()
```

#### Age Distribution

```python
plt.figure(figsize=(15,5))
sns.countplot(x=data['Age'])
plt.xlabel('Age')
plt.show()
```

#### Education Stream Count

```python
plt.figure(figsize=(15,5))
sns.countplot(x=data['Stream'])
plt.xlabel('Stream', fontsize=(10))
plt.ylabel('Count')
plt.title('Education Stream Count')
plt.show()
```

#### Placed Percentage

```python
plt.figure(figsize=(10,5))
labels = ['Not Placed','Placed']
count = [len(data[data['PlacedOrNot'] == 0]), len(data[data['PlacedOrNot'] == 1])]
plt.title('Placed percentage')
plt.pie(count, labels=labels, autopct='%1.1f%%')
plt.show()
```

#### Students did Internship or Not

```python
plt.figure(figsize=(10,5))
sns.countplot(data['Internships'])
plt.xlabel('Number of Internships')
plt.title('Students having Internships')
plt.show()
```

```python
plt.figure(figsize=(10,5))
labels = ['0','1','2','3']
count = [len(data[data['Internships']==0]), len(data[data['Internships']==1]), len(data[data['Internships']==2]),len(data[data['Internships']==3])]
plt.pie(count, labels=labels, autopct='%1.1f%%')
plt.title('Internship Count')
plt.show()
```

#### CGPA Count

```python
plt.figure(figsize=(10,5))
sns.countplot(data['CGPA'])
plt.title('CGPA Count')
plt.show()
```

#### Students living in hostel or not

```python
plt.figure(figsize=(10,5))
labels = ['Lives in hostel','Lives Offcampus']
count = [len(data[data.Hostel == 1]),len(data[data.Hostel == 0])]
plt.pie(count, labels=labels, autopct='%1.1f%%')
plt.title('People Living in hostel')
plt.show()
```

#### Students having Backlogs

```python
plt.figure(figsize=(10,5))
sns.countplot(data['HistoryOfBacklogs'])
plt.title('Backlogs count')
plt.show()
```

```python
plt.figure(figsize=(10,5))
labels = ['Backlogs','No Backlogs']
count = [len(data[data['HistoryOfBacklogs'] == 1]),len(data[data['HistoryOfBacklogs'] == 0])]
plt.pie(count, labels=labels, autopct='%1.1f%%')
plt.title('Students having Backlogs')
plt.show()
```

#### Students who got placed with respective to Age

```python
plt.figure(figsize=(15,5))
sns.histplot(x=data[data['PlacedOrNot']==1]['Age'])
plt.title('Students who got placed with respective to Age')
plt.show()
```

#### Student Stream Count with Placed or Not

```python
plt.figure(figsize=(15,5))
sns.countplot(data['Stream'], hue=data['PlacedOrNot'])
plt.title('Student Stream Count with Placed or Not')
plt.show()
```

#### Gender Count with Placed or Not

```python
plt.figure(figsize=(10,5))
sns.countplot(data['Gender'], hue=data['PlacedOrNot'])
plt.title('Gender Count with Placed or Not')
plt.show()
```

#### Students with Intership got placed or not

```python
plt.figure(figsize=(10,5))
sns.countplot(data['Internships'], hue=data['PlacedOrNot'])
plt.title('Students with Intership got placed or not')
plt.show()
```

```python
plt.figure(figsize=(10,5))
labels = ['Placed','Not placed']
plt.title('Placements < 2 internships')
data2 = data[data.Internships < 2]
count = [len(data2[data2['PlacedOrNot'] == 1]),len(data2[data2['PlacedOrNot'] == 0])]
plt.pie(count, labels = labels, autopct='%1.1f%%')
plt.show()
```

```python
plt.figure(figsize=(10,5))
labels = ['Placed','Not placed']
plt.title('Placements > 2 internships')
data2 = data[data.Internships >= 2]
count = [len(data2[data2['PlacedOrNot'] == 1]),len(data2[data2['PlacedOrNot'] == 0])]
plt.pie(count, labels = labels, autopct='%1.1f%%')
plt.show()
```

#### Students with CGPA got Placed or Not

```python
plt.figure(figsize=(10,5))
sns.countplot(data['CGPA'], hue=data['PlacedOrNot'])
plt.title('Students with CGPA got Placed or Not')
plt.show()
```

### Students who got placed or not,  do they live in hostel or not

### Affect of Backlogs on Placement

```python
sns.countplot(x=data['HistoryOfBacklogs'], hue=data['PlacedOrNot'])
plt.title('Affect of Backlogs on Placement')
plt.show()
```

#### Branchwise placement percentage for Males

```python
data_Male = data[data.Gender == 'Male']
for i in data.Stream.unique():
    labels = ['Placed','Not placed']
    plt.title(i)
    data_Male2 = data_Male[data_Male.Stream == i]
    count = [len(data_Male2[data_Male2.PlacedOrNot == 1]),len(data_Male2[data_Male2.PlacedOrNot == 0])]
    plt.pie(count, labels = labels,autopct='%1.1f%%')
    plt.show()
```

#### Branchwise placement percentage for Females

```python
data_Female = data[data.Gender == 'Female']
for i in data.Stream.unique():
    labels = ['Placed','Not placed']
    plt.title(i)
    data_Female2 = data_Female[data_Female.Stream == i]
    count = [len(data_Female2[data_Female2.PlacedOrNot == 1]),len(data_Female2[data_Female2.PlacedOrNot == 0])]
    plt.pie(count, labels = labels,autopct='%1.1f%%')
    plt.show()
```
